```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib
matplotlib.rc("font", family = "Malgun Gothic")
matplotlib.rc("axes", unicode_minus = False)

from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

import warnings
warnings.filterwarnings("ignore")
pd.options.display.max_columns = 5000
pd.options.display.max_rows = 1000
```

```python
train = pd.read_excel("Downloads/정리.xlsx", header = 0, )
print(train.shape)
train.head(2)

test = pd.read_excel("Downloads/정리.xlsx", header = 0, sheet_name = "Sheet6")
print(test.shape)
test.head(2)
```

```python
train["입고확정일"] = pd.to_datetime(train["입고확정일"])
test["입고확정일"] = pd.to_datetime(test["입고확정일"])

train["year"] = train["입고확정일"].dt.year
train["month"] = train["입고확정일"].dt.month
train["day"] = train["입고확정일"].dt.day
train["week"] = train["입고확정일"].dt.dayofweek

test["year"] = test["입고확정일"].dt.year
test["month"] = test["입고확정일"].dt.month
test["day"] = test["입고확정일"].dt.day
test["week"] = test["입고확정일"].dt.dayofweek
```

```python
feature_names = ['거래액(취소포함)', '취소거래액', '거래액(취소제외)', '상품금액(취소제외)', '배송비(취소제외)',
                 '포인트금액(취소제외)', '쿠폰금액(취소제외)', '실결제금액(취소제외)', '주문건수 (취소제외)', '객단가',
                 '입고 확정 수량', '폐기수량', '폐기 총매입가', 'year', 'month', 'day', 'week']
label_name = '총 매입가'

train_x = train[feature_names]
train_y = train[label_name]
```

```python
from sklearn.model_selection import train_test_split

x_train, y_train, x_test, y_test = train_test_split(train_x, train_y, test_size = 0.15)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### RandomForestRegressor

```python
from sklearn import set_config
set_config(print_changed_only = False)
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRFRegressor, XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

model = RandomForestRegressor()
```

```python
search = 200
hyperparameters_list = []

for i in range(search):
    
    n_estimators = np.random.randint(1000, 3000)
    max_features = np.random.uniform(0, 1)
    max_depth=np.random.randint(low=50,high=100)
    min_samples_split = np.random.uniform(0, 1)
   # min_impurity_split = np.random.uniform(0, 1)
    
    ###
    model = RandomForestRegressor(n_jobs = -1,
                                  random_state = 37,n_estimators = n_estimators, 
                                  max_features = max_features, 
                                  max_depth = max_depth, 
                                  min_samples_split = min_samples_split,
                                  criterion='mse')
    
    ##
    model.fit(train_x, train_y)
    y_predict = model.predict(train_x)
    score=mean_absolute_error(train_y, y_predict)
    
    hyperparameters={"score":score,"n_estimators":n_estimators,"max_depth":max_depth,"max_features":max_features, 
                     "min_samples_split" : min_samples_split}
    hyperparameters_list.append(hyperparameters)
    
    print(f"score={score:.3f},n_estimators={n_estimators},max_depth={max_depth:.6f},max_features={max_features:.6f}, min_samples_split = {min_samples_split:.6f}")
    
pd.DataFrame(hyperparameters_list).sort_values(by = "score", ascending = True).head(5)
```

### GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor()
```

```python
feature_names = ['거래액(취소포함)', '취소거래액', '거래액(취소제외)', '상품금액(취소제외)', '배송비(취소제외)',
                 '포인트금액(취소제외)', '쿠폰금액(취소제외)', '실결제금액(취소제외)', '주문건수 (취소제외)', '객단가',
                 '폐기수량', '폐기 총매입가', 'year', 'month', 'week']
label_name = '총 매입가'


train_x = train[feature_names]
train_y = train[label_name]


search = 200
hyperparameters_list = []

for i in range(search):
    
    n_estimators = np.random.randint(1000, 3000)
    learning_rate = 10 ** -np.random.uniform(1, 2)
    max_features = np.random.uniform(0, 1)
    max_depth=np.random.randint(low=1,high=100)
   # min_samples_split = np.random.randint(1, 10)
    subsample = np.random.uniform(0.5, 1)
   # min_impurity_split = np.random.uniform(0, 1)
    
    ###
    model = GradientBoostingRegressor(n_estimators = n_estimators, 
                                      learning_rate = learning_rate,
                                      max_features = max_features,
                                      subsample = subsample,
                                      max_depth = max_depth)
    
    ##
    model.fit(train_x, train_y)
    y_predict = model.predict(train_x)
    score=mean_absolute_error(train_y, y_predict)
    
    hyperparameters={"score":score,
                     "n_estimators":n_estimators,
                     "learning_rate" : learning_rate,
                     "max_depth":max_depth,
                     "subsample" : subsample,
                     "max_features":max_features}
    hyperparameters_list.append(hyperparameters)
    
    print(f"score={score:.3f},n_estimators={n_estimators},learning_rate = {learning_rate:.3f}, max_depth={max_depth:.6f}, max_features={max_features:.6f}, subsample={subsample:.3f}")
    
pd.DataFrame(hyperparameters_list).sort_values(by = "score", ascending = True).head(5)
```

```python
test_x = test[feature_names]
test_y = test[label_name]


model = GradientBoostingRegressor(n_estimators = 2162, 
                                  learning_rate = 0.090763,
                                  max_features = 0.233110,
                                  subsample = 0.720114,
                                  max_depth = 17)

model.fit(train_x, train_y)
prediction = model.predict(test_x)
prediction
```

```python
test["총 매입가2"] = prediction
test.to_excel("Downloads/결과값4.xlsx")
```
